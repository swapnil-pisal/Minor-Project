## Student Marks Analyzer

Problem Statement -:
1) Student Mark Anylizes
2) Store the marks
3) Calculate -:
     Total Marks , Average Marks , Hieghest marks , Lowest Marks
4) Grade -: On the basis of Average Marks

Logic -:
1) Mark can be stored in the numpy array

2) To find the Total,Average,Heighest,Lowest Marks using numpy methods

3) Grade can find using conditional statement if elif else

```python
# numpy library import
# it is use for numrical python operation
import numpy as np   # np is short and readable

# Marks can be stored in numpy array
marks=np.array([98,76,68,84,43,56,93,87])
print(f"The marks of Student are {marks}")   # f string print statement

# Calculate
# Total Marks
total=np.sum(marks)
print(f"The Total marks of student is {total}")

# Average Marks
average=np.mean(marks)
print(f"The Average marks of student is {average}")

# Heighest Marks
heighest=np.max(marks)
print(f"The Heighest marks of student is {heighest}")

# Lowest Marks
lowest=np.min(marks)
print(f"The Lowest marks of student is {lowest}")

# Grade
# Assign grade based on average
# it using conditional Statement if elif else statement

# if Average marks greater than 85 then grade "A"
if average>85:
    print("Grade A")

# if Average marks greater than or equal to 70 and less than or equal to 85 then grade "B"
elif average<=85 and average>=70:
    print("Grade B")

# if Average marks greater than or equal to 50 and less than 70 then grade "C"
elif average<70 and average>=50:
    print("Grade C")

# if Aaverage marks less than 50 then fail
elif average<50:
    print("Fail")

# invalid Average Marks
else:
    print("Invalid Input")
```
