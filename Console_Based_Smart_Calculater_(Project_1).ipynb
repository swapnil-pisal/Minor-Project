## Console-Based Smart Calculator

Problem Statement -:
1) Perform the multiple mathematical operation
2) Caluculate the Addition,Substraction,Multiplication,Division,Power,Modules are the perform operation using python

Logic Explanation -:

1)The user will enter the calculater. continue or exit there are two option yes or no we use while loop becuase condition is false the out of loop eg.if user says yes then continue and user says no then exit the calculater.

2)After continueing the calculater then show the operations-Addition,Substraction,Multiplication,Division,Power,Modules using the list with each operation is it's own choice

3)eg.1 for addition , 2 for substraction

4)User can select the choice which can operation is required.

5)Then input from the user, user  can enter the two input number using the user input

6)use the if elif else condition for each choice perform each operation


```python
# Asking to the user will use calculater continue or not
# take the user input function
# ask is the variable  for asking to the user
ask=input("Do you want to continue to use this calculater, replay with yes or no: ")

# use while loop use for codition become true until block executed otherwise out of while loop
while ask == "yes":

  # print the the statement if user says yes
  print("Yes! you can continue")

  #  Menu with operations shows
  # menu is variable
  menu=["1 for Addition",
      "2 for Substraction",
      "3 for Multiplication",
      "4 for Division",
      "5 for Power Calculation",
      "6 for Modulus Operation",
     ]    # it is list

  # Display the menu with operation
  print(menu)

  # Use the  user input for choose the choice
  # choice is the variable
  choice=int(input("Enter your choice: "))       # int - for integer number

  # Take the two input from user for the number
  num1 = int(input("Enter the first number: "))     # first number
  num2 = int(input("Enter the second number: "))      # second number

  # use condition statement
  # if-elif-else
  # when the condition is true the execute that block
  # Choose 1 then Addition Operation
  if choice==1:
    print("Addition of two number is: ",num1+num2)  # addition

  # Choose 2 then Substraction Operation
  elif choice==2:
    print("Substraction of two number is: ",num1-num2)  # substraction

  # Choose 3 then Multiplication Operation
  elif choice==3:
    print("Multiplication of two number is: ",num1*num2)    # multiplication

  # Choose 4 then Division Operation
  elif choice==4:
    # cannot divide by zero use condional statement
    if num2==0:
      print("Division by zero is not possible")
    else:
      print("Division of two number is: ",num1/num2)  # division

  # Choose 5 then Power Calculation Operation
  elif choice==5:
    print("Power of two number is: ",num1**num2)   # power

  # Choose 6 then Modules Operation
  elif choice==6:
    print("Modulus of two number is: ",num1%num2)   # reminder

  # Choice is Wrong then else block executed
  else:
    print("Invalid choice")

  # ask it is use for repetation block until true
  ask=input("Do you want to continue to use this calculater, replay with yes or no: ")


# when block false then execute this statement
print("Exit,Thanks for using this calculater ")



```
